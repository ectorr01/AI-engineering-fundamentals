```python
# Setup — eseguite questa cella per prima
!pip install anthropic -q

from google.colab import userdata
import anthropic, os

os.environ["ANTHROPIC_API_KEY"] = userdata.get("ANTHROPIC_API_KEY")
client = anthropic.Anthropic()

def chiedi_claude(messaggio, temperature=0.7, system=None, max_tokens=500):
    params = {
        "model": "claude-haiku-4-5-20251001",
        "max_tokens": max_tokens,
        "temperature": temperature,
        "messages": [{"role": "user", "content": messaggio}]
    }
    if system:
        params["system"] = system
    return client.messages.create(**params).content[0].text

print("✅ Setup completato!")
```

### Esercizio 4 — System prompt e lunghezza della risposta *(libero)*

Aggiungete al system prompt istruzioni sulla lunghezza e il formato
delle risposte. Verificate che Claude le rispetti:

- Versione A: risposte brevi, max 2 frasi
- Versione B: risposte strutturate con bullet point
- Versione C: risposte con sezione 'In breve:' e sezione 'Dettaglio:'

```python
# Esercizio 4 — controllare il formato dell'output
# Questo esercizio è completamente libero

domanda_formato = "Come funziona la vostra piattaforma Xplore?"

# TODO: create 3 versioni del system prompt con istruzioni di formato diverse
# Testate con la stessa domanda
# Claude rispetta le istruzioni di formato?

system_breve = " risposte brevi massimo 2 frasi , ricordati che sei un assistente di widata"  # max 2 frasi
system_bullet = "risposte in formato bullet point ricordati che sei un assistente di widata"  # bullet point
system_strutturato = "risposte brevi ma dettagliate,ricordati che sei un assistente di widata"  # In breve + Dettaglio

for system in [system_breve, system_bullet, system_strutturato]:
    print(f"\n{'='*50}")
    print(f"System prompt: {system}")
    print('='*50)
    print(chiedi_claude(domanda_formato,system=system))
# Conclusione:
# Il system prompt è affidabile per controllare il formato?
# Ci sono casi in cui Claude non rispetta le istruzioni?
# ...
```
